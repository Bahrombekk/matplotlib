```python
import matplotlib.pyplot as plt
from math import sin,cos,degrees,radians,e
```

```python
plt.rcParams['xtick.bottom'] = False
plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['ytick.left'] = False
plt.rcParams['ytick.labelleft'] = False
plt.rcParams['axes.edgecolor'] = 'white'

plt.rcParams['axes.grid'] = True
```

```python
def sine_graph(degree):
  """
  Draw a graph of the sine function.
  
  Args:
    int: degree
  Return:
    None: The sine graph is drawn
  """
  sinus=[]
  x=[]
  degreee=[]
  plt.grid()
  for i in range(degree):
    x.append(i)
    sinus.append(sin(radians(i)))
  plt.scatter(x,sinus,c='#db6440',s=2)
  plt.plot([0,degree],[0,0],c='0.4',lw=2,ls='--',marker='o',mfc='k',markersize=8)
  degreee=[g for g in range(0,degree+1,90)]
  for j in degreee:
    plt.plot([j,j],[0.05,-0.05],c='k')
    plt.text(j,-0.2,f'{j}',ha='center')
  plt.grid()
  plt.show()
```

```python
plt.figure(dpi=150)
sine_graph(720)
```

```python
plt.figure(dpi=150)
sine_graph(180)
```

```python
def sine_and_cosine_graph(degree):
  """
  Draw a graph of the sine and cosine function.
  
  Args:
    int: degree
  Return:
    None: The sine graph is drawn
  """
  sinus=[]
  cosinus=[]
  x=[]
  degreee=[]
  for i in range(degree):
    x.append(i)
    sinus.append(sin(radians(i)))
    cosinus.append(cos(radians(i)))
  plt.scatter(x,sinus,c='#db6440',s=2)
  plt.scatter(x,cosinus,c='#db6440',s=2)
  plt.plot([0,degree],[0,0],c='0.4',lw=2,ls='--',marker='o',mfc='k',markersize=8)
  
  degreee=[g for g in range(0,degree+1,90)]
  for j in degreee:
    plt.plot([j,j],[0.05,-0.05],c='k',)
    plt.text(j,-0.2,f"{j}")
  #WRITE YOUR CODE HERE
  plt.show()
```

```python
plt.figure(dpi=150)
sine_and_cosine_graph(540)
```
