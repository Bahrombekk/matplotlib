jadvalga nom beramiz

plt.title(nom,pad,fontsize,color) va boshqalar kiradi
nom - bu jadvalning nomi stringda beriladi

pad- bu jadval nomi janvaldan qancha uzoqlikda turishi

fontsize-esa nomning katta kichigligi

```python
def nuqta(x,y,color):
  plt.title("Point",pad=10,fontsize=20,color='b')
  for i in range(len(x)):
    plt.scatter(x,y,color='r')
    plt.text(x[i]+0.05,y[i]-0.05,f"({x[i]},{y[i]})")
x=[1,3,5,7]
y=[2,4,6,8]
nuqta(x,y,color='r')
```

Jadvalni o'qlariga nom beramiz

plt.xlabel()- bu x o'qiga nom beradi

plt.ylabel()-bu yigirik o'qiga nom beradi

labelpad- bu nomning jadvaldan qancha uzoqdaligini belgilash bu faqat x va y label da ishlaydi

bungayam colorini berish mumkin va sizini

```python
def nuqta(x,y,color):
  plt.title("Point",pad=10,fontsize=20,color='b')
  plt.xlabel("X o'qi`",labelpad=10,color='b',size=20)
  plt.ylabel("Y o'qi ",labelpad=10,color='b',size=20)
  for i in range(len(x)):
    plt.scatter(x,y,color='r')
    plt.text(x[i]+0.05,y[i]-0.05,f"({x[i]},{y[i]})")
x=[1,3,5,7]
y=[2,4,6,8]
nuqta(x,y,color='r')
```

```python
def subplot_2x3(fig):
  for i in range(1,7):
    plt.subplot(3,2,i)
    plt.text(0.5,0.5,f'{i}',c='r',size=20)
    plt.text(0.4,1.05,f'axis-{i}',size=10,c='b')
  return 0
plt.figure(dpi=50)
plt.show()
```

                                            Sin,cos funksiyalar

```python
from math import sin,cos,pi,radians
```

```python
def sinus1(n):
  sinus=[]
  x=[]
  l=[]
  plt.grid()
  for i in range(n):
    x.append(i)
    sinus.append(sin(radians(i)))
  plt.plot(x,sinus,c='r')
  plt.plot([0,n],[0,0],c='0.4',lw=3,ls='--',marker='o',mfc='k',markersize=8)
  for a in range(0,n+1,90):
    l.append(a)
  for j in l:
    plt.plot([j,j],[0.05,-0.05],c='k')
    plt.text(j,-0.2,f"{j}")
plt.figure(dpi=150)
sinus1(720)
```

```python
sinus1(180)
```

```python
def cosinus1(n):
  cosinus=[]
  x=[]
  l=[]
  plt.grid()
  for i in range(n):
    x.append(i)
    cosinus.append(cos(radians(i)))
  plt.plot(x,cosinus,c='r')
  plt.plot([0,n],[0,0],c='0.4',lw=3,ls='--',marker='o',markersize=8,mfc='k')
  for a in range(0,n+1,90):
    l.append(a)
  for j in l:
    plt.plot([j,j],[0.05,-0.05],c='k')
    plt.text(j,-0.2,f"{j}")
cosinus1(720)
```

```python
def sinus_cosinus(n):
  sinus=[]
  cosinus=[]
  x=[]
  l=[]
  plt.grid()
  for i in range(n):
    x.append(i)
    sinus.append(sin(radians(i)))
    cosinus.append(cos(radians(i)))
  plt.plot(x,sinus,c='r')
  plt.plot(x,cosinus,c='g')
  plt.plot([0,n+1],[0,0],c='0.4',marker='o',mfc='k',markersize=8)
  for a in range(0,n+1,90):
    l.append(a)
  for j in l:
    plt.plot([j,j],[0.05,-0.05],c='k')
    plt.text(j,-0.2,f"{j}")
  plt.legend(['Sin','cos'],ncol=2)
sinus_cosinus(720)
```
